# Exemplo: 

```python
#imports
import math
import matplotlib.pyplot as plt
```

```python
#instanciao
x = [10, 5, 8.5, 2, 1, 1.5, 0, 0.1] #amplitudes
N = len(x) #Amostras
```

```python
#funo do valor de K
def c_val(k):
    if k == 0:
        return math.sqrt(1/2)
    else: 
        return 1

#funo DTC
def dct(x,N):
    #Iniciando array multidimensional
    v = [[0 for col in range(N)] for row in range(N)]
    for k in range(N):
        ck = c_val(k)
        print(f" >>>--- Fazendo para k = {k} ---<<< ")
        for n in range(N):
            v[k][n] = round(math.sqrt(2/N)*ck*x[k]*math.cos(((2*math.pi*n*k)/(2*N))+(k*math.pi)/(2*N)),4)
        
        plt.plot(v[k],'bo')
        plt.ylabel("Pontos para k = "+str(k))
        plt.show()
    return v
```

```python
v = dct(x,N)
```

```python
j = 0
for i in v:
    print(f" Para k = {j}, temos: {i}")
    j+=1
```

```python
soma = [0]*N
for i in range(len(v)):
    for j in range(N):
        soma[j] = round(soma[j]+v[i][j],4)
        
    plt.plot(soma,'bo')
    plt.ylabel("Soma dos pontos de 0 a "+str(i))
    plt.show()

print(" VALOR DA SOMA = ",soma)
plt.plot(soma,'bo')
plt.ylabel("Soma dos Pontos")
plt.show()
```
